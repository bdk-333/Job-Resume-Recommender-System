```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.decomposition import TruncatedSVD

from sklearn.metrics import accuracy_score, classification_report, mean_squared_error
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
utility_matrix = pd.read_csv('../data/processed/utility_matrix.csv', index_col='resume_id')
print("Utility matrix loaded successfully.")
```

### Splitting data:

```python
ratings_array = utility_matrix.values
non_nan_coords = np.argwhere(~np.isnan(ratings_array))
train_coords, test_coords = train_test_split(non_nan_coords, test_size=0.2, random_state=42)
train_matrix = np.full(ratings_array.shape, np.nan)
for r, c in train_coords:
    train_matrix[r, c] = ratings_array[r, c]
test_set = { (r, c): ratings_array[r, c] for r, c in test_coords }
print(f"\nData split into {len(train_coords)} training ratings and {len(test_coords)} testing ratings.")
```

### Tuning SVD:

```python
global_mean_rating = np.nanmean(train_matrix)
train_matrix_filled = np.nan_to_num(train_matrix, nan=global_mean_rating)
item_means = np.nanmean(train_matrix, axis=0)
item_means[np.isnan(item_means)] = global_mean_rating
matrix_demeaned = train_matrix_filled - item_means

# Test a range of values for n_components
# The number of components cannot be larger than the number of features (jobs)
max_k = utility_matrix.shape[1] - 1 # max is number of jobs - 1
k_values = [2, 5, 8, 10, 15, 20, 25, max_k]
rmse_scores = []
```

```python
print("Testing different numbers of latent features (k)...")
for k in k_values:
    # Train SVD model
    svd = TruncatedSVD(n_components=k, random_state=42)
    resume_feature_matrix = svd.fit_transform(matrix_demeaned)
    job_feature_matrix = svd.components_
        
    # Evaluate on the test set
    predictions = []
    actuals = []
    
    for (r, c), actual_rating in test_set.items():
        predicted_rating = np.dot(resume_feature_matrix[r, :], job_feature_matrix[:, c]) + item_means[c]
        predictions.append(predicted_rating)
        actuals.append(actual_rating)
            
    rmse = np.sqrt(mean_squared_error(actuals, predictions))
    rmse_scores.append(rmse)
    print(f"k = {k:2d}  |  Test RMSE = {rmse:.4f}")
```

```python
# Plot the results to find the optimal k
plt.figure(figsize=(10, 6))
plt.plot(k_values, rmse_scores, marker='o', linestyle='--')
plt.title('SVD Performance vs. Number of Latent Features (k)')
plt.xlabel('Number of Latent Features (n_components)')
plt.ylabel('Test Set RMSE:')
plt.grid(True)
plt.show()
```

```python
# Find the best k
best_k = k_values[np.argmin(rmse_scores)]
best_rmse = min(rmse_scores)
print(f"\nOptimal number of latent features (best k) found: {best_k} (with RMSE = {best_rmse:.4f})")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import ast
import re
import os

# --- Scikit-learn Imports ---
from sklearn.model_selection import train_test_split
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics import accuracy_score, classification_report, mean_squared_error
from sklearn.preprocessing import LabelEncoder

from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity

import xgboost as xgb
```

```python
df = pd.read_csv('../data/processed/final_model_data.csv')
df.columns = df.columns.str.strip().str.replace(r'\ufeff', '', regex=True)
if 'responsibilities' in df.columns and 'responsibilities.1' in df.columns:
    if df['responsibilities'].equals(df['responsibilities.1']):
        df.drop(columns=['responsibilities'], inplace=True)
print("Data loaded and initial cleaning complete.")
```

```python
df.columns
```

```python
df1 = df.drop(columns = ['address', 'company_urls', 'start_dates', 'end_dates', 'locations', 'extra_curricular_activity_types',
                        'extra_curricular_organization_links', 'online_links'])
```

```python
df1.columns
```

```python
# Create unique IDs
df1['job_identifier'] = df1[['job_position_name', 'educationaL_requirements', 'experience_years_required', 'age_requirement', 'responsibilities.1', 'skills_required']].astype(str).agg(''.join, axis=1)
df1['resume_identifier'] = df1[['career_objective', 'skills', 'highest_education_level', 'degree_type', 'major_field_of_studies', 'related_skils_in_job', 'positions',
                             'role_positions']].astype(str).agg(''.join, axis=1)
df1['resume_id'] = pd.factorize(df1['resume_identifier'])[0]
df1['job_id'] = pd.factorize(df1['job_identifier'])[0]
```

```python
df1.head()
```

```python
df1.shape
```

## Creating Utility matrix:

```python
# Create and prepare the utility matrix
utility_matrix = df1.pivot_table(index='resume_id', columns='job_id', values='matched_score')
full_matrix_filled = utility_matrix.fillna(utility_matrix.mean().mean())
full_item_means = full_matrix_filled.mean(axis=0).values
full_matrix_demeaned = full_matrix_filled.values - full_item_means
```

```python
utility_matrix.head()
```

### Training SVD:

```python
# Train SVD with our optimal k=2
svd_final = TruncatedSVD(n_components=2, random_state=42)
final_resume_features = svd_final.fit_transform(full_matrix_demeaned)
final_job_features = svd_final.components_
```

```python
all_svd_preds = np.dot(final_resume_features, final_job_features) + full_item_means
svd_preds_df = pd.DataFrame(all_svd_preds, columns=utility_matrix.columns, index=utility_matrix.index)
```

```python
# Un-pivot the matrix to get the SVD prediction for each resume-job pair
svd_preds_long = svd_preds_df.stack().reset_index()
svd_preds_long.columns = ['resume_id', 'job_id', 'svd_predicted_score']
print("SVD feature generated.")
```

### Trying to merge the data:

```python
hybrid_df = pd.merge(df1, svd_preds_long, on=['resume_id', 'job_id'])
print("Hybrid dataset created successfully.")
```

```python
hybrid_df.head()
```

```python
hybrid_df.shape
```

```python
hybrid_df.columns
```

```python
# Define our full feature set
numerical_features = [
    'experience_years_required', 'Skill_Overlap_Count', 'Skill_Jaccard_Score', 
    'Resume_Years_Exp', 'Experience_Mismatch', 'Embedding_Cosine_Similarity', 
    'svd_predicted_score' # <-- The powerful new hybrid feature
]
categorical_features = [
    'Resume_Cluster_KMeans', 
    'highest_education_level', 
    'degree_type'
]

features = numerical_features + categorical_features
```

```python
# hybrid_df.to_csv('../data/processed/hybrid.csv')
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ('num', 'passthrough', numerical_features),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features)
    ], 
    remainder='drop' # This ensures only our selected features are used
)
```

```python
# Define Target
hybrid_df['match_category'] = hybrid_df['matched_score'].apply(lambda s: 'High' if s > 0.7 else ('Medium' if s > 0.4 else 'Low'))
X = hybrid_df[features]
y = hybrid_df['match_category']
y_num = hybrid_df['matched_score']
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

# for regression
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X, y_num, test_size=0.2, random_state=42, stratify=y_encoded)
```

```python
from sklearn.pipeline import Pipeline
model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', xgb.XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='mlogloss'))
])

reg_model_pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('classifier', xgb.XGBRegressor(random_state=42, use_label_encoder=False, eval_metric='mlogloss'))
])
```

```python
# Fit the pipeline
model_pipeline.fit(X_train, y_train)
```

```python
y_pred = model_pipeline.predict(X_test)
```

```python
print("\n--- Final Hybrid Model Performance ---")
print(f"Hybrid Model Test Accuracy: {accuracy_score(y_test, y_pred):.4f}")
print("\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

```python
from sklearn.model_selection import cross_val_score

scores = cross_val_score(model_pipeline, X_train, y_train, cv=5, scoring='accuracy')
print(scores)
```

```python
def create_feature_matrix(new_resume_dict, all_jobs_df):
    """ Generates the complete feature matrix for a new resume against all jobs. """
    print(f"\nProcessing new resume and preparing {len(all_jobs_df)} candidate pairs...")
    resume_df = pd.DataFrame([new_resume_dict])
    candidate_df = pd.merge(resume_df.assign(key=1), all_jobs_df.assign(key=1), on='key').drop('key', axis=1)

    # --- Engineer all required features ---
    
    # Extract Job Experience
    def extract_job_experience(text):
        if isinstance(text, str):
            match = re.search(r'(\d+)', text.lower()); return int(match.group(1)) if match else 0
        return 0
    candidate_df['experience_years_required'] = candidate_df['experiencere_requirement'].apply(extract_job_experience)
    
    # Calculate Experience Mismatch
    candidate_df['Experience_Mismatch'] = abs(candidate_df['Resume_Years_Exp'] - candidate_df['experience_years_required'])
    
    # Engineer Education Features
    def process_degree_data(degree_string):
        master_keywords = ['master', 'm.sc', 'msc', 'm.a', 'mba', 'm.com']; bachelor_keywords = ['bachelor', 'b.sc', 'bsc', 'b.tech', 'bba', 'b.a', 'b.com']; doctorate_keywords = ['ph.d', 'phd', 'doctorate']
        stem_keywords = ['science', 'tech', 'eng', 'computer', 'math', 'statistic']; business_keywords = ['business', 'bba', 'mba', 'account', 'financ', 'commerc']; arts_keywords = ['arts', 'humanities']
        highest_level, degree_type = 'Other', 'Other'
        try:
            degrees = ast.literal_eval(str(degree_string));
            if not isinstance(degrees, list): degrees = [degrees]
            found_levels, found_types = [], []
            for degree in degrees:
                degree_lower = str(degree).lower()
                if any(k in degree_lower for k in doctorate_keywords): found_levels.append(3)
                elif any(k in degree_lower for k in master_keywords): found_levels.append(2)
                elif any(k in degree_lower for k in bachelor_keywords): found_levels.append(1)
                if any(k in degree_lower for k in stem_keywords): found_types.append('STEM')
                elif any(k in degree_lower for k in business_keywords): found_types.append('Business')
                elif any(k in degree_lower for k in arts_keywords): found_types.append('Arts')
            if found_levels:
                max_level = max(found_levels);
                if max_level == 3: highest_level = 'Doctorate'
                elif max_level == 2: highest_level = 'Masters'
                elif max_level == 1: highest_level = 'Bachelors'
            if 'STEM' in found_types: degree_type = 'STEM'
            elif 'Business' in found_types: degree_type = 'Business'
            elif 'Arts' in found_types: degree_type = 'Arts'
        except: pass
        return pd.Series([highest_level, degree_type])
    candidate_df[['highest_education_level', 'degree_type']] = candidate_df['degree_names'].apply(process_degree_data)

    # Use placeholders/defaults for features that require complex pre-trained models
    candidate_df['Embedding_Cosine_Similarity'] = 0.5 
    candidate_df['svd_predicted_score'] = 0.5 
    candidate_df['Skill_Overlap_Count'] = 0
    candidate_df['Skill_Jaccard_Score'] = 0.0
    candidate_df['Resume_Cluster_KMeans'] = 0

    print("Feature matrix created successfully.")
    return candidate_df
```
