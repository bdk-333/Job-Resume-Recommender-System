### SVD: Learning latent details

```python
# Puting  inplace of NaN

utility_matrix_filled = utility_matrix.fillna(0)
```

```python
# standardizing data

matrix_demeaned = utility_matrix_filled.values - np.mean(utility_matrix_filled.values, axis=0)
print("\nUtility matrix prepared for SVD (filled NaNs and demeaned).")
```

```python
# Using Truncated SVD as factorizer

from sklearn.decomposition import TruncatedSVD
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
ratings_array = utility_matrix.values
non_nan_coords = np.argwhere(~np.isnan(ratings_array))
train_coords, test_coords = train_test_split(non_nan_coords, test_size=0.2, random_state=42)
train_matrix = np.full(ratings_array.shape, np.nan)
for r, c in train_coords:
    train_matrix[r, c] = ratings_array[r, c]
test_set = { (r, c): ratings_array[r, c] for r, c in test_coords }
print(f"\nData split into {len(train_coords)} training ratings and {len(test_coords)} testing ratings.")
```

```python
global_mean_rating = np.nanmean(train_matrix)
train_matrix_filled = np.nan_to_num(train_matrix, nan=global_mean_rating)
item_means = np.nanmean(train_matrix, axis=0)
item_means[np.isnan(item_means)] = global_mean_rating
matrix_demeaned = train_matrix_filled - item_means

# Test a range of values for n_components
# The number of components cannot be larger than the number of features (jobs)
max_k = utility_matrix.shape[1] - 1 # max is number of jobs - 1
k_values = [2, 5, 8, 10, 15, 20, 25, max_k]
rmse_scores = []
```

```python
print("Testing different numbers of latent features (k)...")
for k in k_values:
    # Train SVD model
    svd = TruncatedSVD(n_components=k, random_state=42)
    resume_feature_matrix = svd.fit_transform(matrix_demeaned)
    job_feature_matrix = svd.components_
        
    # Evaluate on the test set
    predictions = []
    actuals = []
    
    for (r, c), actual_rating in test_set.items():
        predicted_rating = np.dot(resume_feature_matrix[r, :], job_feature_matrix[:, c]) + item_means[c]
        predictions.append(predicted_rating)
        actuals.append(actual_rating)
            
    rmse = np.sqrt(mean_squared_error(actuals, predictions))
    rmse_scores.append(rmse)
    print(f"k = {k:2d}  |  Test RMSE = {rmse:.4f}")
```

```python
# Plot the results to find the optimal k
plt.figure(figsize=(10, 6))
plt.plot(k_values, rmse_scores, marker='o', linestyle='--')
plt.title('SVD Performance vs. Number of Latent Features (k)')
plt.xlabel('Number of Latent Features (n_components)')
plt.ylabel('Test Set RMSE:')
plt.grid(True)
plt.show()
```

## Creating a train-test split

```python
from sklearn.model_selection import train_test_split
```

```python
ratings_array = utility_matrix.values
    
# Get the coordinates of all non-NaN ratings
non_nan_coords = np.argwhere(~np.isnan(ratings_array))
    
# Split these coordinates into training and testing sets
train_coords, test_coords = train_test_split(non_nan_coords, test_size=0.2, random_state=42)

# Create the training matrix with NaNs for the test set ratings
train_matrix = np.full(ratings_array.shape, np.nan)
    
for r, c in train_coords:
    train_matrix[r, c] = ratings_array[r, c]

# Create the test set dictionary
test_set = { (r, c): ratings_array[r, c] for r, c in test_coords }
print(f"\nData split into {len(train_coords)} training ratings and {len(test_coords)} testing ratings.")
```

```python
# Filling nan in training data with global mean (important for factorization)

global_mean_rating = np.nanmean(train_matrix)

train_matrix_filled = np.nan_to_num(train_matrix, nan=global_mean_rating)
    
# De-mean the data by subtracting the item (job) means
item_means = np.nanmean(train_matrix, axis=0)
    
# Handle jobs with no ratings in the training set
item_means[np.isnan(item_means)] = global_mean_rating
matrix_demeaned = train_matrix_filled - item_means
print("\nTraining matrix prepared for SVD.")
```

```python
print("\n--- Applying TruncatedSVD to discover latent features ---")
    
n_components = 2
svd = TruncatedSVD(n_components=n_components, random_state=42)
    
resume_feature_matrix = svd.fit_transform(matrix_demeaned)
job_feature_matrix = svd.components_
print(f"SVD complete. Discovered {n_components} latent features.")
```

### Making predictions on test set, and calculating RMSE:

```python
from sklearn.metrics import root_mean_squared_error, mean_squared_error
```

```python
print("\n--- Evaluating model accuracy with RMSE ---")
    
predictions = []
actuals = []
    
for (r, c), actual_rating in test_set.items():
    # Reconstruct the predicted rating for this specific cell
    predicted_rating = np.dot(resume_feature_matrix[r, :], job_feature_matrix[:, c]) + item_means[c]
    predictions.append(predicted_rating)
    actuals.append(actual_rating)
        
rmse = np.sqrt(mean_squared_error(actuals, predictions))
print(f"Model RMSE on the test set: {rmse:.4f}")
```

```python
# filling missing values with mean
full_matrix_filled = utility_matrix.fillna(utility_matrix.mean().mean())
full_item_means = full_matrix_filled.mean(axis=0).values

# standardizing
full_matrix_demeaned = full_matrix_filled.values - full_item_means

# final svd object, with n=2, as it was the best
svd_final = TruncatedSVD(n_components=n_components, random_state=42)

# making matrix
final_resume_features = svd_final.fit_transform(full_matrix_demeaned)
final_job_features = svd_final.components_
all_predicted_ratings = np.dot(final_resume_features, final_job_features) + full_item_means

predicted_ratings_df = pd.DataFrame(all_predicted_ratings, columns=utility_matrix.columns, index=utility_matrix.index)
print("\n--- Final model trained on all data and full predicted ratings matrix created. ---")
```

### function to recommend jobs for a resume:

```python
def recommend_jobs_for_resume(resume_id, reconstructed_matrix, top_n=5):
    """Recommends the top N jobs for a given resume_id."""
    if resume_id not in reconstructed_matrix.index:
        return f"Resume ID {resume_id} not found."
    
    # Get the row of predicted scores for the target resume
    resume_ratings = reconstructed_matrix.loc[resume_id]
    top_jobs = resume_ratings.sort_values(ascending=False).head(top_n)
    return top_jobs
```

```python
print("\n--- Example: Getting recommendations for Resume ID 10 ---")
recommendations = recommend_jobs_for_resume(10, predicted_ratings_df, top_n=5)
print("Top 5 Recommended Jobs for Resume 10:")
print(recommendations)
```

## Creating traditional Collaborative filtering:

```python
# --- Create Train-Test Split ---
ratings_array = utility_matrix.values
non_nan_coords = np.argwhere(~np.isnan(ratings_array))
train_coords, test_coords = train_test_split(non_nan_coords, test_size=0.2, random_state=42)
train_matrix_df = utility_matrix.copy()

# "Hide" the test ratings by setting them to NaN in the training matrix
for r, c in test_coords:
    train_matrix_df.iloc[r, c] = np.nan
    
test_set = { (r, c): ratings_array[r, c] for r, c in test_coords }
print(f"\nData split into training and testing sets.")
```

```python
from sklearn.metrics.pairwise import pairwise_distances
```

```python
mean_rating = np.nanmean(train_matrix_df.values)
user_similarity = 1 - pairwise_distances(train_matrix_df.fillna(mean_rating), metric='cosine')
```

```python
def predict_user_based(user_id, item_id, train_data, user_sim_matrix):
        """Predicts a rating using user-based CF."""
        # Get similarities for the active user, excluding the user itself
        sim_scores = user_sim_matrix[user_id].copy()
        sim_scores[user_id] = 0 # Set self-similarity to 0
        
        # Get the ratings for the target item from all other users
        item_ratings = train_data.iloc[:, item_id]
        
        # Select users who have rated this item
        rated_users = item_ratings.notna()
        
        # If no other users have rated this item, predict the mean
        if not rated_users.any():
            return mean_rating
            
        # Get the similarity scores of only the users who have rated the item
        relevant_sim_scores = sim_scores[rated_users]
        # Get the ratings of only the users who have rated the item
        relevant_ratings = item_ratings[rated_users]
        
        # Calculate the weighted average
        weighted_sum = np.dot(relevant_sim_scores, relevant_ratings)
        sum_of_weights = relevant_sim_scores.sum()
        
        if sum_of_weights == 0:
            return mean_rating
        
        return weighted_sum / sum_of_weights
```

```python
# Make predictions for the test set
user_cf_predictions = []
actuals = []

for (r, c), actual_rating in test_set.items():
    user_cf_predictions.append(predict_user_based(r, c, train_matrix_df, user_similarity))
    actuals.append(actual_rating)
        
rmse_user_cf = np.sqrt(mean_squared_error(actuals, user_cf_predictions))
print(f"User-Based CF Test RMSE: {rmse_user_cf:.4f}")
```

With different distance metrics, we get the following RMSE:

1. Cosine: 0.1407
2. Manhattan: 0.1529
3. Euclidean: 29.5104

### Creating Item-based collaborative filtering (recommend resumes based on jobs):

```python
item_similarity = 1 - pairwise_distances(train_matrix_df.T.fillna(mean_rating), metric='euclidean')
```

```python
def predict_item_based(user_id, item_id, train_data, item_sim_matrix):
        """Predicts a rating using item-based CF."""
        # Get similarities for the active item, excluding the item itself
        sim_scores = item_sim_matrix[item_id].copy()
        sim_scores[item_id] = 0
        
        # Get the ratings for the target user across all items
        user_ratings = train_data.iloc[user_id, :]
        
        # Select items that this user has rated
        rated_items = user_ratings.notna()
        
        if not rated_items.any():
            return mean_rating
            
        # Get similarity scores and ratings for only the rated items
        relevant_sim_scores = sim_scores[rated_items]
        relevant_ratings = user_ratings[rated_items]
        
        # Calculate the weighted average
        weighted_sum = np.dot(relevant_sim_scores, relevant_ratings)
        sum_of_weights = relevant_sim_scores.sum()
        
        if sum_of_weights == 0:
            return mean_rating
            
        return weighted_sum / sum_of_weights
```

```python
# Make predictions for the test set
item_cf_predictions = []
for (r, c), actual_rating in test_set.items():
    item_cf_predictions.append(predict_item_based(r, c, train_matrix_df, item_similarity))
        
rmse_item_cf = np.sqrt(mean_squared_error(actuals, item_cf_predictions))
print(f"Item-Based CF Test RMSE: {rmse_item_cf:.4f}")
```

Different distances and their RMSE:

1. Cosine: 0.1529
2. Manhattan: 0.1691
3. Euclidean: 0.1727

### Function to find similar jobs:

```python
def recommend_similar_jobs(job_id, reconstructed_matrix, top_n=5):
    """
    Finds jobs similar to a given job_id based on the reconstructed ratings.
    It does this by correlating the predicted ratings for the target job with all other jobs.
    """
    if job_id not in reconstructed_matrix.columns:
        return f"Job ID {job_id} not found."
        
    # Get the column of predicted ratings for our target job
    target_job_ratings = reconstructed_matrix[job_id]
    
    # Correlate this column with all other columns in the matrix
    similar_to_target_job = reconstructed_matrix.corrwith(target_job_ratings)
    
    # Create a DataFrame for easy sorting
    corr_df = pd.DataFrame(similar_to_target_job, columns=['correlation'])
    corr_df.dropna(inplace=True)
    
    # Sort by correlation and get the top N
    # We drop the first one because it will be the target job itself (correlation of 1.0)
    top_recommendations = corr_df.sort_values('correlation', ascending=False).head(top_n + 1)
    top_recommendations = top_recommendations.iloc[1:]
    
    return top_recommendations
```

```python
print("\n--- Example: Getting recommendations for Job ID 0 ---")
    
# Let's get recommendations for the first job (job_id = 0)
recommendations = recommend_similar_jobs(0, predicted_ratings_df, top_n=5)
    
print("Jobs most similar to Job 0:")
print(recommendations)
```

```python
completed_matrix_df = utility_matrix.copy()
    
# Find all cells that were originally empty (or we hid for testing)
missing_coords = np.argwhere(np.isnan(completed_matrix_df.values))
    
print(f"Found {len(missing_coords)} missing ratings to predict and fill.")
    
# Loop through the missing coordinates and fill with predictions
# We use the item-based model as it often performs better.
for r, c in missing_coords:
    predicted_rating = predict_item_based(r, c, train_matrix_df, item_similarity)
    completed_matrix_df.iloc[r, c] = predicted_rating
        
print("\nSuccessfully filled all missing values. The matrix is now complete.")
print("Displaying the head of the completed utility matrix:")
display(completed_matrix_df.head())
```
