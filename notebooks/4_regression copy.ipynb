```python
def weighted_tfidf(weight):
    return Pipeline([
        ('tfidf', TfidfVectorizer(max_features=100, stop_words='english')),
        ('scale', FunctionTransformer(lambda x: x * weight, accept_sparse=True))
    ])
```

```python
weighted_transformers = [
    ('skills', weighted_tfidf(3), 'skills'),
    ('skills_required', weighted_tfidf(3), 'skills_required'),
    ('responsibilities.1', weighted_tfidf(2), 'responsibilities.1'),
    ('career_objective', weighted_tfidf(1), 'career_objective'),
    ('degree_names', weighted_tfidf(1), 'degree_names'),
    ('professional_company_names', weighted_tfidf(1), 'professional_company_names'),
    ('educationaL_requirements', weighted_tfidf(1), 'educationaL_requirements'),
    ('﻿job_position_name', weighted_tfidf(1), '﻿job_position_name'),
]
```

```python
weighted_model_df = df[[col for _, _, col in weighted_transformers] + ['matched_score']].copy()
weighted_model_df.fillna('', inplace=True)
```

# Roberta Embeddings

```python
from transformers import RobertaTokenizer, RobertaForSequenceClassification, Trainer, TrainingArguments
from torch.utils.data import Dataset
```

```python
tokenizer = RobertaTokenizer.from_pretrained("roberta-base")
```

```python
def tokenize_data(df, max_length=256):
    input_texts = df["input_text"].astype(str).tolist()
    encodings = tokenizer(
        input_texts,
        max_length=max_length,
        padding=True,
        truncation=True,
        return_tensors="pt"
    )
    labels = torch.tensor(df["label"].values, dtype=torch.float)
    return encodings, labels
```
